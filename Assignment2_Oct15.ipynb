#### 4. Filter the rows of iris_2d that has petallength (3rd column) > 1.5 and sepallength (1st column) < 5.0

```python
iris = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data' 

iris_2d = np.genfromtxt(iris, delimiter=',', dtype='float', usecols=[0,1,2,3])

filt = iris_2d[(iris_2d[:, 2] > 1.5) & (iris_2d[:, 0] < 5.0)]

print(filt)
```

#### 7. How to get the rows of a dataframe with row sum > 100?

```python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))

# axis = 1 sets it to find the sum of rows 
filt_df = df[df.sum(axis=1) >100]

filt_df
```
